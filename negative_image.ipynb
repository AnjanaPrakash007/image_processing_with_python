```python
from PIL import Image,ImageOps
def print_image(image_path):
    img=Image.open(image_path)
    display(img)
    negative_img=ImageOps.invert(img)
    display(negative_img)
    greyscale_image=ImageOps.grayscale(img)
    display(greyscale_image)
    negative_grey_image=ImageOps.invert(greyscale_image)
    display(negative_grey_image)
image_path="images/odns.jpg"
print_image(image_path)
```
