```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("images/bird.webp")
row, column ,_= img.shape
img1=np.zeros((row,column),dtype='uint8')
min_range=100
max_range=200
for i in range(row):
    for j in range(column):
        if min_range<img[i,j,0]<max_range:
            img1[i,j]=255
        else:
            img1[i,j]=0
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.imshow(img)
plt.title("original image")
plt.subplot(122)
plt.imshow(img1,cmap="gray")
plt.title("sliced image")
plt.tight_layout()
plt.show()
```
