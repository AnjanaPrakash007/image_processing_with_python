```python
from PIL import Image,ImageOps
def print_image(image_path):
    img=Image.open(image_path)
    display(img)
    negative_image=ImageOps.invert(img)
    display(negative_image)
    gray_scale_image=ImageOps.grayscale(img)
    display(gray_scale_image)
    negative_gray_scale_image=ImageOps.invert(gray_scale_image)
    display(negative_gray_scale_image)
   
    
image_path="images/bird.webp"
print_image(image_path)
```
