```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy import fftpack
img=cv2.imread("images/bird.webp",0)
fft2d=fftpack.fft2(img)
fft2d_shifted=fftpack.fftshift(fft2d)
magnitude_spectrum_2d=np.log(np.abs(fft2d_shifted)+1)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(img,cmap="gray")
plt.title("Original image")
plt.subplot(132)
plt.imshow(magnitude_spectrum_2d,cmap="gray")
plt.title("2D fourier Transform")
plt.subplot(133)
plt.plot(np.fft.fftfreq(img.shape[1]),np.log(np.abs(fftpack.fftshift(fftpack.fft(img,axis=1))).mean(axis=0)+1),label='rows')
plt.plot(np.fft.fftfreq(img.shape[0]),np.log(np.abs(fftpack.fftshift(fftpack.fft(img,axis=1))).mean(axis=0)+1),label='columns')
plt.title('combined 1D fourier transforms')
plt.xlabel('frequency')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
# Read the image in grayscale mode
img = cv2.imread("images/bird.webp",0)
# Compute the 2D Fourier transform
fft2d = fftpack.fft2(img)
# Shift the zero-frequency component to the center
fft2d_shifted = fftpack.fftshift(fft2d)
# Compute the magnitude spectrum of the shifted 2D Fourier transform
magnitude_spectrum_2d = np.log(np.abs(fft2d_shifted) + 1)
col_fft = np.log(np.abs(fftpack.fftshift(fftpack.fft(img, axis=0))) + 1)
row_fft = np.log(np.abs(fftpack.fftshift(fftpack.fft(img, axis=1))) + 1)
# Compute and plot the 1D Fourier transform along rows
plt.figure(figsize=(12, 6))
plt.subplot(141)
plt.imshow(img, cmap="gray")
plt.title("Original image")
plt.subplot(142)
plt.imshow(magnitude_spectrum_2d, cmap="gray")
plt.title("2D Fourier Transform")
plt.subplot(143)
plt.imshow(row_fft, cmap="gray")
plt.title("1D Fourier Transform (Rows)")
# Compute and plot the 1D Fourier transform along columns
plt.subplot(144)
plt.imshow(col_fft, cmap="gray")
plt.title("1D Fourier Transform (Columns)")
plt.tight_layout()
plt.show()
```
