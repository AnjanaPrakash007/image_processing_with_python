```python
import cv2
import matplotlib.pyplot as plt
#reading image
img=cv2.imread("images/drmdm.jpg")
resized_image=cv2.resize(img,(200,200))
rotated_image=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
flip_horizontal=cv2.flip(img,1)
flip_vertical=cv2.flip(img,0)
gray_scale_image=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blurred_image=cv2.GaussianBlur(img,(3,3),0)
# plotting
plt.figure(figsize=(12,6))
plt.subplot(241)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("original image")
plt.subplot(242)
plt.imshow(cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB))
plt.title("Resized image")
plt.subplot(243)
plt.imshow(cv2.cvtColor(rotated_image,cv2.COLOR_BGR2RGB))
plt.title("Rotated image")
plt.subplot(244)
plt.imshow(cv2.cvtColor(flip_horizontal,cv2.COLOR_BGR2RGB))
plt.title("Horizontaly flipped ")
plt.subplot(245)
plt.imshow(cv2.cvtColor(flip_vertical,cv2.COLOR_BGR2RGB))
plt.title("verticaly flipped")
plt.subplot(246)
plt.imshow(cv2.cvtColor(blurred_image,cv2.COLOR_BGR2RGB))
plt.title("blurred image")
plt.subplot(247)
plt.imshow(gray_scale_image,cmap="gray")
plt.title("Gray scale image")
plt.tight_layout()
plt.show()
```
