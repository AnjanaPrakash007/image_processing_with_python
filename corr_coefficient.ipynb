```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img1=cv2.imread("images/drmdm.jpg")
img2=cv2.imread("images/kannur1.jpg")
plt.figure(figsize=(12,6))
plt.subplot(131)
plt.imshow(img1)
plt.title("First image")
plt.subplot(132)
plt.imshow(img2)
plt.title("Second image")
img1=cv2.resize(img1,(img2.shape[1],img2.shape[0]))
plt.subplot(133)
plt.imshow(img1)
plt.title("Resized image")
mean_img1=np.mean(img1)
std_img1=np.std(img1)
mean_img2=np.mean(img2)
std_img2=np.std(img2)
cor_coeff=np.corrcoef(img1.flatten(),img2.flatten())[0,1]
print("Mean of first image",mean_img1)
print("Mean of second image",mean_img2)
print("standard deviation of first image",std_img1)
print("standard dev of second image",std_img2)
print("correlation coefficient",cor_coeff)
```
