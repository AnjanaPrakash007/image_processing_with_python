```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("images/low_contrast.jpg",0)
min_intensity=np.min(img)
max_intensity=np.max(img)
stretched_image=(((img-min_intensity)/(max_intensity-min_intensity))*255).astype(np.uint8)
equalised_image=cv2.equalizeHist(img)
hist_equalised_image=cv2.calcHist([equalised_image],[0],None,[255],[0,255])
hist_streched_image=cv2.calcHist([stretched_image],[0],None,[255],[0,255])
hist_original_image=cv2.calcHist([img],[0],None,[255],[0,255])
plt.figure(figsize=(12,6))
plt.subplot(231)
plt.imshow(img)
plt.title("original image")
plt.subplot(232)
plt.imshow(stretched_image)
plt.title("stretched image")
plt.subplot(233)
plt.imshow(equalised_image)
plt.title("equalised image")
plt.subplot(234)
plt.plot(hist_original_image)
plt.title("histogram for image")
plt.xlim([0,256])
plt.subplot(235)
plt.plot(hist_streched_image)
plt.title("histogram for  streched image")
plt.xlim([0,256])
plt.subplot(236)
plt.plot(hist_equalised_image)
plt.title("histogram for equalised image")
plt.xlim([0,256])
plt.tight_layout()
plt.show()
```
